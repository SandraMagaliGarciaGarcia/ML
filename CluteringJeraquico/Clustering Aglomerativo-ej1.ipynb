```python
# Import the necessary librarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# datos generados aleatoriamente en Python, crenando 50 muestras y dos caracteristicas utilizando la funcion random de Numpy
np.random.seed(0)
x = np.random.randn(50, 2)
```

```python
# Graficar el dendograma utilizando su función 
z = linkage(x, method='complete')
 
# Plot the dendrogram
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
dendrogram(z)
plt.show()
```
