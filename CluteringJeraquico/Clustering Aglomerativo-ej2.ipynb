```python
# Import the necessary librarias
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import matplotlib.pyplot as plt
```

```python
# datos generados aleatoriamente en Python, crenando 10 muestras y 5 caracteristicas utilizando la funcion random de Numpy
np.random.seed(0)
X = np.random.randn(20, 2)
```

```python
Z = linkage(X, method='ward', metric='euclidean')
```

```python
# Plot dendrogram
# Plot the dendrogram
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
dendrogram(Z)
plt.show()
```
